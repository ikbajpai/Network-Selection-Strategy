```python
import random as rd

n = int(input("Please Enter thr number of available networks: "))  # specify the number of networks

networks = []
for i in range(n):
    network = {
        "name": f"Network {i+1}",
        "technology": rd.choice(["3G", "4G", "5G"]),
        "bandwidth": rd.uniform(1, 30),
        "energy_efficiency": rd.uniform(0, 1),
        "delay_jitter": rd.uniform(0, 2),
        "price": rd.uniform(20, 150)
    }
    networks.append(network)
# print(networks,"\n")
for network in networks:
    print(network)
    print()
```

```python
# Calculate the score for each network
for network in networks:
    score = (network["bandwidth"] / network["price"]) * network["energy_efficiency"] / (network["delay_jitter"] + 1)
    network["score"] = score

# Sort the networks by score in descending order
networks = sorted(networks, key=lambda x: x["score"], reverse=True)

# Print the sorted list of networks with their scores
for network in networks:
    print(network["name"], "\tScore:", network["score"])
```

```python
# Python 3 program to calculate Distance Between Two Points on Earth
import math
from math import radians, cos, sin, asin, sqrt
def distance(lat1, lon1, lat2, lon2):

	# The math module contains a function named
	# radians which converts from degrees to radians.
	lon1 = radians(lon1)
	lon2 = radians(lon2)
	lat1 = radians(lat1)
	lat2 = radians(lat2)

	# Haversine formula
	dlon = lon2 - lon1
	dlat = lat2 - lat1
	a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2

	c = 2 * asin(sqrt(a))

	# Radius of earth in kilometers. Use 3956 for miles
	r = 6371

	lat2 = math.degrees(lat2)
	lon2 = math.degrees(lon2)

	# calculate the result
	return(c * r,(lat2, lon2))


# driver code
# lat1 = 53.32055555555556
# lat2 = 53.31861111111111
# lon1 = -1.7297222222222221
# lon2 = -1.6997222222222223
# print(distance(lat1, lat2, lon1, lon2), "K.M")
```

```python
import random as rd
def Base_Station_Coordinates():
    random_latitude = rd.uniform(15.47310943904706, 15.312142624568104)
    random_longitude = rd.uniform(74.92777295698296, 75.23695008278617)
    return(random_latitude, random_longitude)
```

```python
no_of_BS_in_Globe = 200
BSList=[]
for i in range(no_of_BS_in_Globe):
    BSList.append(Base_Station_Coordinates())
```
